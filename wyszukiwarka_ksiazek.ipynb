```python
from bs4 import BeautifulSoup
import requests
from time import sleep
import re
import pandas as pd
from pandas import DataFrame, Series
```

```python
base_url = 'https://helion.pl/kategorie/big-data'
soup = BeautifulSoup(requests.get(base_url).text,'html5lib')
```

```python
books= []

num_pages = 5

analizowana_klasa = soup('ul', 'book-type book-type-points')
print(len(analizowana_klasa)) #zwraca liczbę książek wyświetlanych na stronie

def is_video(ul):            #chcemy analizować tylko książki
    popups = ul('span','popup')
    return (len(popups) == 1 and popups[0].text.strip() == 'Videokurs')

print(len([ul for ul in analizowana_klasa if not is_video(ul)]))

def book_info(div):
    title = div.find('div','book-info-middle').a.text
    author = div.find('p','author').text
    cena = soup.find(class_='price price-incart').a.text
    return {'tytuł':title,'autorzy':author, 'cena':cena}

for page_num in range(1,num_pages+1):
    print('souping page', page_num, ',', len(books), 'found so far')
    url = base_url + str(page_num)

    for div in soup('div', 'book-info'):
        if not is_video(div):
            books.append(book_info(div))
```
